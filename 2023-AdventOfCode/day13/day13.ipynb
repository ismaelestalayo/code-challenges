```python
def find_mirroring_row_start(pattern):
    hashes = []
    for line in pattern:
        hashes.append(hash(line))

    hashes_zipped = list(zip(hashes, hashes[1:]))

    repeating_indexes = [i for i, (h1, h2) in enumerate(hashes_zipped) if (h1 == h2)]
    if (len(repeating_indexes) == 0):
        # print("❌❌❌ No repeating pattern found ❌❌❌")
        pass
    else:
        for repeating_index in repeating_indexes:
            repeating_index = repeating_index + 1

            pattern_up = pattern[:repeating_index]
            pattern_down = pattern[repeating_index:]
            min_len = min(len(pattern_up), len(pattern_down))

            pattern_up_match = pattern_up[-min_len:]
            pattern_down_match = pattern_down[:min_len]

            if (pattern_up_match == pattern_down_match[::-1]):
                return repeating_index
    return 0

def find_mirroring_col_start(pattern):
    # Transpose the 2D list
    pattern = [''.join(row) for row in zip(*pattern)]
    return find_mirroring_row_start(pattern)
```

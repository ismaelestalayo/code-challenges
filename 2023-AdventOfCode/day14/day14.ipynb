```python
import re


lines = open("example1.txt").read()
lines = open("input.txt").read()

lines = lines.split("\n")

# ############################################################################
# print(lines[10:])

dish = [[c for c in line] for line in lines]
dish[:10]
```

```python
def transpose(matrix):
    # Transpose the 2D list
    return list(row for row in zip(*matrix))
```

```python
# transpose
transposed = transpose(dish)
_ = [print(line) for line in transposed]
```

```python
line = dish[3]

print("".join(line))

def move_rocks_left(line):
    chunks = "".join(line).split("#")
    new_chunks = []
    new_chunks = ["".join(sorted(chunk)[::-1]) for chunk in chunks]
    new_chunks = "#".join(new_chunks)
    return [nc for nc in new_chunks]

new_line = move_rocks_left(line)
```

```python
# Tranpose once
transposed = transpose(dish)

# Move rock to the left
_ = [print(t) for t in dish]
for i, line in enumerate(transposed):
    transposed[i] = move_rocks_left(line)

# Tranpose the matrix back, and its like the rocks moved up
transposed = transpose(transposed)
transposed = transpose(transposed)
transposed = transpose(transposed)

print()
_ = [print("".join(t)) for t in transposed]
```

### Main code

```python
transposed = transpose(dish)

for i, line in enumerate(transposed):
    transposed[i] = move_rocks_left(line)

transposed = transpose(transposed)
transposed = transpose(transposed)
transposed = transpose(transposed)


num_rows = len(transposed)
SOLUTION = 0
for i, line in enumerate(transposed):
    SOLUTION += line.count("O") * (num_rows - i)

print(f"SOLUTION: {SOLUTION}")
```

# Part 2
The rocks roll **north**, then **west**, then **south**, then **east**

1.000.000.000 times!!!!  (1000000000)

```python
import re


lines = open("input.txt").read()
lines = open("example1.txt").read()

lines = lines.split("\n")

# ############################################################################
# print(lines[10:])

dish = [[c for c in line] for line in lines]
dish[:10]
```

```python
line = dish[3]

print("".join(line))

def move_rocks(line, invert: bool = False):
    chunks = "".join(line).split("#")
    new_chunks = []
    if (invert):
        new_chunks = ["".join(sorted(chunk)[::-1]) for chunk in chunks]
    else:
        new_chunks = ["".join(sorted(chunk)) for chunk in chunks]
    new_chunks = "#".join(new_chunks)
    return [nc for nc in new_chunks]


new_line = move_rocks(line, True)
print("".join(new_line))
```

```python
class Cable:    
    def __init__(self, line):
        self.cable, connections = line.split(": ")
        self.conns = connections.strip().split(" ")
        self.cablesConnected = 0

    def __str__(self):
        return f"Cable '{self.cable}' ({self.cablesConnected}) connected to {self.conns}"


Cable(lines[0])
```

```python
cables = [Cable(line) for line in lines]
# bricks = sorted(bricks, key=lambda b: (b.z1))

for i, cable in enumerate(cables):
    cable.cablesConnected = len([c for c in cables if cable.cable in c.conns])

_ = [print(cables.index(cable), cable) for cable in cables]

"""
Cable 0 'jqt' (1) connected to ['rhn', 'xhk',          'nvd']

Cable 3 'cmg' (1) connected to ['qnr', 'nvd', 'lhk',   'bvb']

Cable 6 'pzl' (1) connected to ['lsr',                 'hfx',    'nvd']

"""
```

### Draw the graph

```python
import networkx as nx
import matplotlib.pyplot as plt


def build_graph(cables):
    G = nx.Graph()
    for cable in cables:
        for connection in cable.conns:
            G.add_edge(cable.cable, connection)
    return G

def draw_tree(graph):
    pos = nx.spring_layout(graph)
    nx.draw(graph, pos, with_labels=True, font_weight='bold', node_size=700, node_color='skyblue', font_size=8, font_color='black')
    plt.show()



_ = [print(cable) for cable in cables]

graph = build_graph(cables)
draw_tree(graph)
```

```python
lines = open("input.txt").read().split("\n")

cables = [Cable(line) for line in lines]


graph = build_graph(cables)
draw_tree(graph)
```

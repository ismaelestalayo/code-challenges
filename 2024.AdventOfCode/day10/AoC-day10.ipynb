```python
def parse_file(filename: str):
    text_line = open(filename).read()
    text_lines = text_line.split("\n")
    text_line = [[int(n) for n in line] for line in text_lines]
    return text_line


lines = parse_file("example1.txt")
_ = [print(l) for l in lines]
```

```python
def find_zeroes(lines: list):
    zeroes = []
    for i, line in enumerate(lines):
        for j, num in enumerate(line):
            if num == 0:
                zeroes.append((i, j))
    return zeroes


lines = parse_file("example1.txt")
zeroes = find_zeroes(lines)
zeroes
```

```python
def walk(coord, reached_peaks: list, debug=False):
    x, y = coord
    curr = MAP[x][y]
    up = MAP[x - 1][y] if x > 0 else -2
    down = MAP[x + 1][y] if x < ROWS - 1 else -2
    left = MAP[x][y - 1] if y > 0 else -2
    right = MAP[x][y + 1] if y < COLS - 1 else -2
    if(curr == 9):
        if debug: print(f"> PEAK! (at {x, y}) ")
        reached_peaks.append( (x, y) )
        return reached_peaks
    if (up == curr + 1):
        if debug: print(f"  - Currently at { coord } ({ curr })... â˜")
        walk( (x - 1, y), reached_peaks, debug)
    if (down == curr + 1):
        if debug: print(f"  - Currently at { coord } ({ curr })... ðŸ‘‡ðŸ»")
        walk( (x + 1, y), reached_peaks, debug)
    if (left == curr + 1):
        if debug: print(f"  - Currently at { coord } ({ curr })... ðŸ‘ˆðŸ»")
        walk( (x, y - 1), reached_peaks, debug)
    if (right == curr + 1):
        if debug: print(f"  - Currently at { coord } ({ curr })... ðŸ‘‰ðŸ»")
        walk( (x, y + 1), reached_peaks, debug)
    return reached_peaks


MAP = parse_file("example1.txt")
ROWS, COLS = len(MAP), len(MAP[0])
zeroes = find_zeroes(MAP)
x, y = zeroes[0]
print(f"> Trying with { zeroes[0] }")

reached_peaks = walk( (x, y), [], True)
print( len(set(reached_peaks)) )
```

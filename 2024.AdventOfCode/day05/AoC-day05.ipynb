```python
def extract_rules_and_updates(text_lines):
    index = text_lines.index("")
    text_ordering_rules = text_lines[:index]
    text_updates = text_lines[index+1:]

    # RULES
    rules = [[ int(rule) for rule in ordering_rule.split("|")] for ordering_rule in text_ordering_rules]
    dict_rules = {rule[0]: [] for rule in rules}
    [dict_rules[rule[0]].append(rule[1]) for rule in rules]
    # sort
    for key in dict_rules.keys():
        dict_rules[key].sort()

    # UPDATES
    updates = [[int(u) for u in update.split(",")] for update in text_updates]

    return dict_rules, updates
```

```python
text_lines = open("example1.txt").read().split("\n")
text_lines = open("input.txt").read().split("\n")

dict_rules, updates = extract_rules_and_updates(text_lines)

correct_updates = []
for update in updates:
    # print(update)
    incorrect = False
    for i, page in enumerate(update):
        if page in dict_rules:
            previous = update[:i]
            for prev in previous:
                if prev in dict_rules[page]:
                    # print(f"  ❌ {prev} in {dict_rules[page]}")
                    incorrect = True
                    break
        if incorrect:
            break
    if not incorrect:
        # print("  ✅ CORRECT")
        correct_updates.append(update)

sum([
        correct_update[len(correct_update) // 2 ]
        for correct_update in correct_updates
])
```

```python
text_lines = open("example1.txt").read().split("\n")
dict_rules, updates = extract_rules_and_updates(text_lines)

def check_update(nums):
    incorrect = False
    for i in range(len(nums)):
        page = nums[i]
        if page in dict_rules:
            previous = nums[:i]
            for prev in previous:
                if prev in dict_rules[page]:
                    incorrect = True
                    # swap
                    nums[i], nums[i-1] = nums[i-1], nums[i]
                    i -= 1
                    print(f"   ❌ {page} [{prev}] in {dict_rules[page]} (i = { i })")
    if incorrect:
        return nums
    else:
        return None


fixed_updates = []
for update in updates:
    print("-" * 20)
    print(update)
    fixed = check_update(update.copy())
    if fixed:
        fixed_updates.append(fixed)

sum([
    fixed_update[len(fixed_update) // 2 ]
    for fixed_update in fixed_updates
])
```

```python
def parse_text_line(text_line):
    test, equation = text_line.split(":")
    test = int(test)
    equation = [eq for eq in equation.strip().split(" ")]
    return test, equation

def parse_text_lines(text_lines):
    return [parse_text_line(text_line) for text_line in text_lines]


parse_text_line( text_lines[0] )
```

```python
import itertools

def generate_combinations(num, operators="+*"):
    return [''.join(p) for p in itertools.product(operators, repeat=num)]

combinations = generate_combinations(2)
print(combinations)
```

```python
line = parse_text_line( text_lines[0] )

combinations = generate_combinations( 1 )
for combination in combinations:
    result, nums = line
    equation = list(zip( nums, ["+", "*"], nums[1:]))[0]
    # eq_str = "".join( equation )
    print( equation )
```

```python
def evaluate_equations(equations, debug=False):
    result = equations[0][0]
    for equation in equations:
        if debug: print(f"> Evaluating: {equation}...")
        new_eq = (str(result),) + equation[1:]
        result = eval( "".join(new_eq) )
        if debug: print(f"  Result: {result}")
    return result


print( evaluate_equations( [('10', '+', '19')], True ) )
print( evaluate_equations( [('81', '+', '40'), ('40', '+', '27')], True ) )
```

```python
text_lines = open("example1.txt").read().split("\n")
lines = parse_text_lines( text_lines )

calibration_result = 0
for line in lines:
    test, nums = line
    combinations = generate_combinations( len(nums) - 1 )
    # CALC COMBINATIONS
    for combination in combinations:
        equation = list(zip( nums, combination, nums[1:]))
        result = evaluate_equations( equation )
        # print(f"Eq: {equation} -> {result}")
        if (result == test):
            print(f"FOUND: { equation } == { test }")
            calibration_result += test
            break
    # print("-" * 20)

print(f"Calibration result: { calibration_result }")
```

```python
def evaluate_equations(equations, debug=False):
    result = equations[0][0]
    for equation in equations:
        if debug: print(f"> Evaluating: {equation}...")
        new_eq = (str(result),) + equation[1:]
        if (new_eq[1] != "|"):
            result = eval( "".join(new_eq) )
        else:
            result = int( str(new_eq[0]) + str(new_eq[2]) )
        if debug: print(f"  Result: {result}")
    return result


print( evaluate_equations( [('15', '|', '6')], True ) )
print( evaluate_equations( [('6', '*', '8'), ('8', '|', '6'), ('8', '*', '15')], True ) )
```
